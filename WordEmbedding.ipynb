```python
sentences = [['I', 'love', 'nlp'],
['I', 'will', 'learn', 'nlp', 'in', '2','months'],
['nlp', 'is', 'future'],
[ 'nlp', 'saves', 'time', 'and', 'solves','lot', 'of', 'industry', 'problems'],
['nlp', 'uses', 'machine', 'learning']]
```

```python
import gensim
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```python
#Train the model
sk = Word2Vec(sentences, vector_size =50, window = 3, min_count=1,sg = 0 )
print(sk)
```

```python
print(sk.wv['nlp'])
```

```python
# save the model
sk.save('sk.bin')
```

```python
sk = Word2Vec.load('sk.bin')
```

```python
# PCA is one way to evaluate word embedding
x = [sk.wv[word] for word in sk.wv.index_to_key]
pca = PCA(n_components=2)
result = pca.fit_transform(x)
```

```python
import matplotlib.pyplot as pyplot

# Get the list of words
words = list(sk.wv.index_to_key)

# Plot the PCA results
pyplot.scatter(result[:, 0], result[:, 1])

# Annotate each point with the corresponding word
for i, word in enumerate(words):
    pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))

pyplot.show()
```

```python
cbow = Word2Vec(sentences, vector_size =50, window = 3, min_count=1,sg = 0 )
print(cbow)
# access vector for one word
print(cbow.wv['nlp'])
# save model
cbow.save('cbow.bin')
# load model
cbow = Word2Vec.load('cbow.bin')
# T – SNE plot
x = [cbow.wv[word] for word in cbow.wv.index_to_key]
pca = PCA(n_components=2)
result = pca.fit_transform(x)
# create a scatter plot of the projection
pyplot.scatter(result[:, 0], result[:, 1])
words = list(cbow.wv.index_to_key)
for i, word in enumerate(words):
    pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))
pyplot.show()
```

#Implement FastText

```python
from gensim.models import FastText
from sklearn.decomposition import PCA
from matplotlib import pyplot
```

```python
fast = FastText(sentences,vector_size=20, window=1, min_count=1,workers=5, min_n=1, max_n=2 )
```

```python
print(fast.wv['nlp'])
```

```python
x = [fast.wv[word] for word in fast.wv.index_to_key]
pca=PCA(n_components=2)
result = pca.fit_transform(x)
```

```python
pyplot.scatter(result[:,0],result[:,1])
words  = list(fast.wv.index_to_key)
for i, word in enumerate(words):
  pyplot.annotate(word, xy=(result[i, 0], result[i,1]))
pyplot.show()
```
