```python
# load librares
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import re
```

```python
df = pd.read_csv('sentimentdataset.csv')
```

```python
x = df['Text']
y= df['Sentiment']
```

```python
y.unique()
```

```python
x
```

```python
# preprocessing
def preprocessing_text(text):
  text = re.sub(r'/d+',"",text) # remove the number in the text
  text = text.translate(str.maketrans('','',string.punctuation)) # it can remove the punctuation
  text = text.lower()
  text = ' '.join([word for word in text.split() if word not in stopwords.words('english')])
  lemmatizer = WordNetLemmatizer()
  text  = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
  return text
pre = [preprocessing_text(review) for  review in x]
```

```python
pre
```

```python
# model buliding
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_ve = TfidfVectorizer()
tf = tf_ve.fit_transform(pre)
```

```python
c=pd.DataFrame(tf.toarray())
```
