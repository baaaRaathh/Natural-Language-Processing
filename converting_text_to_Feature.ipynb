#converting text into feature  using Countvertorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = ['i love python and im using pyhton to nlp nad i wiil learn nlp in 2 weeks']
```

```python
vec = CountVectorizer()
```

```python
vec.fit(text)
```

```python
vect = vec.transform(text)
```

```python
print(vec.vocabulary_)
print(vect.toarray())
```

##Bigram using countvectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = ['i love python and im using pyhton to nlp nad i wiil learn nlp in 2 weeks']
```

```python
vec= CountVectorizer(ngram_range= (3,3))
vec.fit(text)
vect = vec.transform(text)
```

```python
print(vec.vocabulary_)
print(vect.toarray())
```

#Hash Vectorizer

```python
from sklearn.feature_extraction.text import HashingVectorizer
```
